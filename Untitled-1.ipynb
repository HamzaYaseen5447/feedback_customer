PREPROCESSING
    Lower Case
    Tokenization(Break words)
    Removing Special Characters
    Removing stop words(like is, of) and punctuations
    Stemming(dance dancing dances = dance bcuz all means dance)

```python
from nltk.corpus import stopwords   
# stopwords.words('english')           ## Have no contribution in the sentence and use for sentence forming
```

```python
import string
string.punctuation
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
ps.stem('running')
```

```python
def text_process(text):
    text = text.lower()
    text = nltk.word_tokenize(text)

    y = []

    for i in text:
        if i.isalnum():
            y.append(i)

    text = y[:] # cloning the list
    y.clear()

    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)

    text = y[:] # cloning the list
    y.clear()

    for i in text:
        y.append(ps.stem(i))

    return ' '.join(y) # return string
```

MODEL BUILDING: Naive Bayes bcuz it works best on textual data

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

X = cv.fit_transform(df1['Ticket Description']).toarray()   # Convert a collection of text documents to a matrix of token counts
y = df1['Sentiment_Score']

print('X is: ', X)
print('y is: ', y)
```

```python
from sklearn.model_selection import train_test_split    
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix
from sklearn.naive_bayes import MultinomialNB, GaussianNB, BernoulliNB

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)

gnb = GaussianNB()  
mnb = MultinomialNB()   
bnb = BernoulliNB()
```

```python
df1.head(5)
```

```python
gnb.fit(X_train, y_train)
y_pred1 = gnb.predict(X_test)
print('GaussianNB using CountVectorizer: ')
print('Accuracy Score: ', accuracy_score(y_test, y_pred1))
print('Precision Score: ', precision_score(y_test, y_pred1, average='weighted'))
print('Confusion Matrix: ', confusion_matrix(y_test, y_pred1))
```

```python
mnb.fit(X_train, y_train)
y_pred2 = mnb.predict(X_test)
print('MultinomialNB using CountVectorizer: ')
print('Accuracy Score: ', accuracy_score(y_test, y_pred2))
print('Precision Score: ', precision_score(y_test, y_pred2, average='weighted'))
print('Confusion Matrix: ', confusion_matrix(y_test, y_pred2))
```

```python
bnb.fit(X_train, y_train)
y_pred3 = bnb.predict(X_test)
print('BernoulliNB using CountVectorizer: ')
print('Accuracy Score: ', accuracy_score(y_test, y_pred3))
print('Precision Score: ', precision_score(y_test, y_pred3, average='weighted'))
print('Confusion Matrix: ', confusion_matrix(y_test, y_pred3))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
def predict_sentiment(user_input, model, vectorizer):
    user_vector = vectorizer.transform([user_input])
    sentiment_score = model.predict(user_vector)[0]

    sentiment_label = {1: 'Positive', 0: 'Neutral', -1: 'Negative'}
    sentiment = sentiment_label.get(sentiment_score, "Unknown")

    return sentiment_score, sentiment
```

```python
import speech_recognition as sr

def support_response(sentiment):
    # Define different responses based on sentiment
    if sentiment == "Negative":
        return "We're sorry to hear about your issue. Our support team will look into this immediately."
    elif sentiment == "Neutral":
        return "Thank you for your feedback. How can we assist you further?"
    elif sentiment == "Positive":
        return "Thank you for your positive feedback! We're glad to know you're satisfied."

def recognize_speech():
    recognizer = sr.Recognizer()
    
    # Use the microphone as source
    with sr.Microphone() as source:
        print("Listening for your issue...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)
    
    try:
        print("Recognizing your speech...")
        # Convert speech to text
        text = recognizer.recognize_google(audio)
        print(f"You said: {text}")
        return text
    except sr.UnknownValueError:
        print("Sorry, I couldn't understand your speech. Please try again.")
        return None
    except sr.RequestError:
        print("Sorry, there was an issue with the speech recognition service.")
        return None

def chatbot():
    print("Welcome to the Customer Support Chatbot!")
    
    # Ask user to choose between typing or speech
    mode_choice = input("Press 1 for Chatting (Typing) or 2 for Speech (Speaking): ")
    
    if mode_choice == "1":
        print("You selected Chat mode. Please type your issue.")
        
        while True:
            user_input = input("\nEnter Ticket Description (or type 'exit' to quit): ")
            
            # Exit if the user types 'exit'
            if user_input.lower() == 'exit':
                print("Goodbye!")
                break
            
            # Print the user input
            print(f"\nYou entered: {user_input}")
            
            # Predict sentiment
            sentiment_score, sentiment = predict_sentiment(user_input, mnb, cv)
            
            # Show sentiment and response
            print(f"Sentiment Score: {sentiment_score} ({sentiment})")
            response = support_response(sentiment)
            print(f"Response: {response}")

    elif mode_choice == "2":
        print("You selected Speech mode. Please speak your issue.")
        
        while True:
            user_input = recognize_speech()
            
            # If no valid speech input, continue to the next iteration
            if not user_input:
                continue
            
            # Exit if the user says 'exit'
            if user_input.lower() == 'exit':
                print("Goodbye!")
                break
            
            # Print the user input
            print(f"\nYou said: {user_input}")
            
            # Predict sentiment
            sentiment_score, sentiment = predict_sentiment(user_input, mnb, cv)
            
            # Show sentiment and response
            print(f"Sentiment Score: {sentiment_score} ({sentiment})")
            response = support_response(sentiment)
            print(f"Response: {response}")
    
    else:
        print("Invalid choice. Exiting...")
        
# Run the chatbot
chatbot()


```
